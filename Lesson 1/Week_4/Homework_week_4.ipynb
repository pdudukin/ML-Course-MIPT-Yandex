###### Часть 1: Выбрано распределение $Chi^{2}$(100)

```python
%matplotlib inline 

# Libraries Import
from scipy.stats import chi2, norm
import matplotlib.pyplot as plt
import numpy as np
import math as m

df = 100 # number of degrees of freedom / 100 is enough for CLT
chi2_rv = chi2.rvs(df, size=1000) # generate sample (n=1000) from Chi(df) distribution
chi2_real = chi2(df) # generate real chi(2)
x = np.arange(50,150,0.1) # generate set of x

# plot graph
plt.figure(figsize=(20,4))
plt.hist(chi2_rv, 50, normed = True, label = 'Actual Chi_sq(100)') # plot histogram 
plt.plot(x, chi2_real.pdf(x), 'k-', linewidth = 3.5, color = 'Red', label = 'Theoretical Chi_sq(100)') # plot Chi_sq(100) pdf
plt.title('n=5')
plt.legend(loc='best', frameon=False) # display legend
```

```python
n = [5, 10, 50] # Вектор - объем выборок
N = 1000 # размер выборок
x = np.arange(75,125,0.1) 
x5 = np.zeros(N)
norm_pdf = norm(df,m.sqrt(2*df/n[0]))

# считаем сумму сгенеренных выборок
for i in range(0,n[0]):
    Chi2_rv5 = chi2.rvs(df, size=N)
    for j in range(0,N):
        x5[j] += Chi2_rv5[j]
        
x_n5 = x5/n[0] # Вычисляем среднее от вектора-суммы

# строим график (для последующих графиков все аналогично)
plt.figure(figsize=(20,4))
plt.hist(x_n5, 75, normed=True, label = 'Actual Gauss distribution') # plot histogram
plt.plot(x, norm_pdf.pdf(x), 'k-', linewidth=3.5, color = 'Red', label='Theoretical Gauss PDF') # plot pdf
plt.title('n=5')
plt.legend(loc='best', frameon=False) # display legend
```

```python
# n = 10, N = 1000
x10 = np.zeros(N)
x = np.arange(80,120,0.1)
norm_pdf = norm(df,m.sqrt(2*df/n[1]))

for i in range(0,n[1]):
    Chi2_rv10 = chi2.rvs(df, size=N)
    for j in range(0,N):
        x10[j] += Chi2_rv10[j]
        
x_n10 = x10/n[1] # Вычисляем среднее
plt.figure(figsize=(20,4))
plt.hist(x_n10, 50, normed=True, label = 'Actual Gauss distribution') # plot histogram
plt.plot(x, norm_pdf.pdf(x), 'k-', linewidth=3.5, color = 'Red', label='Theoretical Gauss PDF') # plot Chi2 pdf
plt.title('n=10')
plt.legend(loc='best', frameon=False) # display legend
```
